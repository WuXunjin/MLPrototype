```python
import torch
import torch.nn as nn

from model_exp0a import Generator, Discriminator, EmbeddingNet

import matplotlib.pyplot as plt

numPrimaryInputChannels = 1
modelFolder = 'dummy'
```

```python
def getWeights(fineName):
    
    E = EmbeddingNet(numPrimaryInputChannels)
    PATH = modelFolder + '/' + EFinalS
    E.load_state_dict(torch.load(PATH)) 
    
    x = []
    for m in E._modules:
        if isinstance(E._modules[m], nn.Conv2d) or isinstance(E._modules[m],nn.BatchNorm2d):
            x.append(E._modules[m].weight)
    return x
```

```python
def plotHist(layerList, number):
    plt.hist(layerList[number].data.numpy().flatten())
    plt.show()
```

### Random Initialization

```python
EFinalS = 'E_random_van.pt'
x=getWeights(EFinalS)
plotHist(x,0)
```

```python
plotHist(x,1)
```

### Triplet Initialised

```python
EFinalS = 'E_triplet_van.pt'
x=getWeights(EFinalS)
plotHist(x,0)
```

```python
plotHist(x,1)
```

### Contrastive Initialised

```python
EFinalS = 'E_contrastive_van.pt'
x=getWeights(EFinalS)
plotHist(x,0)
```

```python
plotHist(x,1)
```
